## Probability density at t=0

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook 
```

```python
plt.style.use("ggplot")
```

```python
import scipy.constants as const
import math
import numpy as np
```

```python
pi = const.pi
hbar = const.hbar
```

```python
p2 = 10**(-34)        # Some random value
p1 = 0.8*10**(-34)
```

```python
x = np.linspace(-30, 30, 600)
```

```python
rho = [2/(pi*(hbar**2)) * (math.cos(((p1 - p2)*x)/hbar))**2 for x in x]  # at t = 0
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, rho);
```

```python
ax.set_xlabel("x");
```

```python
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
# ax.set_xlim(-30, 30)
```

```python
ax.tick_params(labelleft=False, left=False)
```

```python
ax.set_yticklabels([]);
```

```python
ax.set_title("Probability distribution at t = 0");
```

```python
E1 = 10**(-34)
E2 = 0.8*E1
```

```python
from matplotlib import animation
```

```python
fig = plt.figure(figsize=(9, 3.5))
a = animation.FuncAnimation(fig, update, interval=50)
```
